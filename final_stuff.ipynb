```python
import os
import pandas as pd
import librosa
import librosa.display
import glob
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.layers import Dense
from keras.layers import Dropout 
from keras.layers import Activation
from keras.layers import Dropout
from keras.layers import SimpleRNN
from keras.layers import Flatten
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import SimpleRNN
from keras.layers import GRU

ipd.Audio("air_conditioner.wav")
ipd.Audio("car_horn.wav")
ipd.Audio("children_playing.wav")
ipd.Audio("dog_bark.wav")
ipd.Audio("drilling.wav")
ipd.Audio("gun_shot.wav")
ipd.Audio("jackhammer.wav")
ipd.Audio("siren.wav")
ipd.Audio("street_music.wav")

data1,sr1 = librosa.load("air_conditioner.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("air_conditioner")
plt.show()

data1,sr1 = librosa.load("car_horn.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("car_horn")
plt.show()

data1,sr1 = librosa.load("children_playing.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("children_playing")
plt.show()

data1,sr1 = librosa.load("dog_bark.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("dog_bark")
plt.show()

data1,sr1 = librosa.load("drilling.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("drilling")
plt.show()

data1,sr1 = librosa.load("gun_shot.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("gun_shot")
plt.show()

data1,sr1 = librosa.load("jackhammer.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("jackhammer")
plt.show()

data1,sr1 = librosa.load("siren.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("siren")
plt.show()

data1,sr1 = librosa.load("street_music.wav")
plt.figure(figsize = (12,4))
librosa.display.waveplot(data1,sr = sr1)
plt.title("street_music")
plt.show()

data_dir = "/home/aditya/audio_analysis/"

df = pd.read_csv(data_dir + "train.csv")

columns0 = ['Class']
columns1 = ["ID"]
target = pd.DataFrame(df, columns=columns0)
X = pd.DataFrame(df, columns = columns1)

labelencoder = LabelEncoder()
y = np_utils.to_categorical(labelencoder.fit_transform(target))

x = numpy.array(X)

soundfile = []
for i in range(5435):
    soundrecordings = data_dir +"Train/" + str(x[i][0]) + ".wav"
    X_, sample_rate = librosa.load(soundrecordings, res_type='kaiser_fast')
    mfccs = numpy.mean(librosa.feature.mfcc(y=X_, sr=sample_rate, n_mfcc=40).T,axis=0)
    mf = numpy.reshape(mfccs, (1,40))
    soundfile.append(mf)
    
input_X = numpy.concatenate(soundfile, axis=0)
model = Sequential()
model.add(Dense(256, input_dim = 40, activation = "relu"))
model.add(Dense(256, activation = "relu"))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

history = model.fit(input_X, y, epochs = 10, validation_split = 0.2)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

model = Sequential()
model.add(Dense(256, input_dim = 40, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

history = model.fit(input_X, y, epochs = 10)
plt.plot(history.history['acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()

plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()


model = Sequential()
model.add(Dense(256, input_dim = 40, activation = "relu"))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

history = model.fit(input_X, y, epochs = 10, validation_split = 0.05)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

X_forRNN = numpy.reshape(input_X, (5435, 1, 40))

model = Sequential()
model.add(LSTM(128, input_shape = (1,40)))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = "categorical_crossentropy", metrics = ['accuracy'], optimizer = 'adam')

history = model.fit(X_forRNN,y, epochs = 10, validation_split = 0.1)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()


model = Sequential()
model.add(SimpleRNN(128, input_shape = (1,40)))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = "categorical_crossentropy", metrics = ['accuracy'], optimizer = 'adam')

history = model.fit(X_forRNN,y, epochs = 10, validation_split = 0.1)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()

model = Sequential()
model.add(GRU(128, input_shape = (1,40)))
model.add(Dense(10, activation = 'softmax'))

model.summary()

model.compile(loss = "categorical_crossentropy", metrics = ['accuracy'], optimizer = 'adam')

history = model.fit(X_forRNN,y, epochs = 10, validation_split = 0.2)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper_left')
plt.show()
```
